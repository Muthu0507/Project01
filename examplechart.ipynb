```python
# Import Pandas
import pandas as pd

# Upload CSV File
from google.colab import files
uploaded = files.upload()

# Read Dataset
df = pd.read_csv(next(iter(uploaded)))

# Dataset Preview
df.head()
```

```python
print(df.columns)
```

```python
import matplotlib.pyplot as plt

df['attrition_flag'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)
plt.title('Attrition Distribution')
plt.ylabel('')
plt.show()
```

```python
import pandas as pd

dept_attrition = pd.crosstab(
    df['department'],
    df['attrition_flag']
)

dept_attrition.plot(kind='bar')
plt.title('Department-wise Attrition')
plt.xlabel('Department')
plt.ylabel('Employee Count')
plt.show()
```

```python
salary_attrition = pd.crosstab(
    df['salary_band'],
    df['attrition_flag']
)

salary_attrition.plot(kind='bar')
plt.title('Salary Band vs Attrition')
plt.xlabel('Salary Band')
plt.ylabel('Employee Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# 1. Attrition Distribution
df['attrition_flag'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    ax=axes[0,0]
)
axes[0,0].set_title("Attrition Distribution")
axes[0,0].set_ylabel('')

# 2. Department-wise Attrition
pd.crosstab(df['department'], df['attrition_flag']).plot(
    kind='bar',
    ax=axes[0,1]
)
axes[0,1].set_title("Department-wise Attrition")

# 3. Salary Band vs Attrition
pd.crosstab(df['salary_band'], df['attrition_flag']).plot(
    kind='bar',
    ax=axes[1,0]
)
axes[1,0].set_title("Salary Band vs Attrition")

# 4. Job Satisfaction
df['job_satisfaction_score'].plot(
    kind='hist',
    bins=10,
    ax=axes[1,1]
)
axes[1,1].set_title("Job Satisfaction Score")

plt.tight_layout()
plt.show()
```
